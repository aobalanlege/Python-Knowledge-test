## **Task3: Compute the Gini Impurity,entropy and Information Gain (30 points)**

 **(3.1)** - Write a piece of code to compute Gini Impurity, Entropy
and Information Gain for each split.

Save the Gini Impurity in a list named `gini`,
save the Information Gain in a list named `gainInformation`. Note that you are required to implement a function for each, and you are not allowed to use any existing implementation of it in any Python library.
The data that is represented in the code block below are factors that affect whether you would go out to play golf or not. 


```python
import pandas as pd

data = pd.DataFrame([['Rainy','Hot','High','False'],['Rainy','Hot','High','True'],['Overcast','Hot','High','False'],['Sunny','Mild','High','False'],['Sunny','Cool','Normal','False'],['Sunny','Cool','Normal','True'],['Overcast','Cool','Normal','True'],['Rainy','Mild','High','False'],['Rainy','Cool','Normal','False'],['Sunny','Mild','Normal','False'],['Rainy','Mild','Normal','True'],['Overcast','Mild','Hight','True'],['Overcast','Hot','Normal','False'],['Sunny','Mild','Hight','True']], columns=["Outlook","Temperature","Humidity","Windy"])
class_labels = ['No', 'No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
data["Play_Golf"] = class_labels
data
```

Entropy(S) = $-\sum \limits_{i=1}^{n} P_i\times Log_2(P_i)$

Gini impurity = $1 - \sum \limits_{i=1}^{n}(P_i)^2$

information gain(S,A) = $H(S) -\sum \limits_{i=1}^{n}\frac{|(S_i)|}{|(S)|}\times H(S)= H(S)-H(A,S)$

S = Target or label data
A = features or attribute column
H = entropy
H(S,A)

```python
# Code for 3.1

def gini_index_calculate(prob_list):

    impurity = 1
    for item in prob_list:
        impurity -= item ** 2
    return impurity

def entropy_calculate(prob_list):

    entropy = 0
    for item in prob_list:
        entropy -= item * np.log2(item)
    return entropy

conditions,counts = np.unique(data.Play_Golf,return_counts=True)
P = [count/data.shape[0] for count in counts]
print('Probabilities of %s and %s are %.3f, %.3f respectively'%(conditions[0],conditions[1],P[0],P[1]))
entropy_entire = entropy_calculate(P)
gini_entire = gini_index_calculate(P)
print('Entire dataset entropy is %.3f bits'%entropy_entire)
print('Entire dataset gini is %.3f bits'%gini_entire)
```

# **(3.2)** -  Perform the First Split using the Gain Information 

Outlook Split

```python
# Code for 3.2
conditions_Outlook,counts_Outlook= np.unique(data.Outlook,return_counts=True)
P_Outlook = [count/len(data) for count in counts_Outlook]
print('First Split: Outlook:')
for condition, probability in zip(conditions_Outlook,P_Outlook):
    print('\tProbabality of %s is %.3f'%(condition, probability))
    
# Outlook entropy
entropy_Outlook={}
total_entropy_Outlook=0
for condition, probability in zip(conditions_Outlook,P_Outlook):
    conditions,counts = np.unique(data['Play_Golf'][data['Outlook']==condition],return_counts=True)
    P = [count/len(data[data.Outlook==condition]) for count in counts]
    entropy_Outlook[condition]=entropy_calculate(P)
    total_entropy_Outlook += entropy_calculate(P)*probability
    gini_Outlook = gini_index_calculate(P_Outlook)
    information_gain_Outlook = entropy_entire - total_entropy_Outlook
print('\ngini impurity is %.3f bits'% gini_Outlook)
print('\n')
# information gain for Outlook

for condition, entropy in entropy_Outlook.items():
    print('\tEntropy for %s is %.2f'%(condition,entropy))
print('\nEntropy at Outlook is %.3f'%total_entropy_Outlook)
print('\nInformation gain for Outlook is %.3f'%information_gain_Outlook)

```
